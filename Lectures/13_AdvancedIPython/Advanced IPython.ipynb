```python
X = np.arange(100)
Y = 200.*X + 100.
```

```python
%%R -i X,Y -o XYcoeff
XYlm = lm(Y~X)
print(summary(XYlm))
XYcoeff = coef(XYlm)
par(mfrow=c(2,2))
plot(XYlm)
```

### Cython (C and Python mashup)

```python
%load_ext cythonmagic
```

```python
def f(x):
    return x**2-x

def integrate_f(a, b, N):
    s = 0
    dx = (b-a)/N
    for i in range(N):
        s += f(a+i*dx)
    return s * dx
```

```python
%%cython

cdef double f(double x):
    return x**2-x

def cyintegrate_f(double a, double b, int N):
    cdef int i
    cdef double s, dx
    s = 0
    dx = (b-a)/N
    for i in range(N):
        s += f(a+i*dx)
    return s * dx
```

```python
a, b, N = 0, 2, 10000
%timeit integrate_f(a, b, N)
%timeit cyintegrate_f(a, b, N)
```

## 3. Interactive Widgets

**NOTE** Only in IPython 2.0 and later!

The IPython website has [documentation and examples](http://nbviewer.ipython.org/github/ipython/ipython/blob/2.x/examples/Interactive%20Widgets/Index.ipynb)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from IPython.html.widgets import interactive
from IPython.display import Audio, display
```

```python
def beat_freq(f1=220.0, f2=224.0):
    max_time = 3
    rate = 8000
    times = np.linspace(0,max_time,rate*max_time)
    signal = np.sin(2*np.pi*f1*times) + np.sin(2*np.pi*f2*times)
    print(f1, f2, abs(f1-f2))
    display(Audio(data=signal, rate=rate))
    return signal
```

```python
v = interactive(beat_freq, f1=(200.0,300.0), f2=(200.0,300.0))
display(v)
```

```python
v.kwargs
```

```python
f1, f2 = v.children
f1.value = 255
f2.value = 260
plt.plot(v.result[0:6000])
```
